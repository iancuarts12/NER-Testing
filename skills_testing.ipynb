```python
import spacy
import json

nlp = spacy.load("en_core_web_md")
jsonl_path = "skill_patterns.jsonl" 

def load_skill_patterns(jsonl_path):
    patterns = []
    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line.strip())
            patterns.append(data)
    return patterns

patterns = load_skill_patterns(jsonl_path)
ruler = nlp.add_pipe("entity_ruler", name="skill_ruler", before="ner")
ruler.add_patterns(patterns)

def add_skill_patterns(nlp, patterns):
    ruler = nlp.add_pipe("entity_ruler", before="ner")
    ruler.add_patterns(patterns)
    return ruler

def test_skill_extraction(sentence, patterns_path):
    patterns = load_skill_patterns(patterns_path)
    ruler = add_skill_patterns(nlp, patterns)

    doc = nlp(sentence)

    skills = []
    for ent in doc.ents:
        if ent.label_ == "SKILLS":
            skills.append(ent.text)

    return skills

test_sentences = [
    "Experience plumber with expertisee in toilet installation, drian unclogging, and pipe repair.:",
    "Skilleed in electrical wiriing, conduit installation, and LED light installation.",
    "Appliance maintenance specialist for washing machine, blenders, and rice cookers.",
    "Furniture repair expert with wood polishing and cushion replacement skills."
]

skill_pattern_path = "skill_patterns.jsonl"

print("Testing Skill Extraction:")
print("="*50)

for i, sentence in enumerate(test_sentences, 1):
    extracted_skills = test_skill_extraction(sentence, skill_pattern_path)
    print(f"\nTest {i}:")
    print(f"Sentence: {sentence}")
    print(f"Extracte Skills: {extracted_skills}")
    print("-"*30)
```

```python
import spacy
import json

nlp = spacy.load("en_core_web_md")
jsonl_path = "skill_patterns.jsonl"
services_jsonl_path= "services_patterns.jsonl" 

def load_service_patterns(services_jsonl_path):
    patterns = []
    with open(services_jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line.strip())
            patterns.append(data)
        return patterns

def load_skill_patterns(jsonl_path):
    patterns = []
    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line.strip())
            patterns.append(data)
    return patterns

service_patterns = load_service_patterns(services_jsonl_path)
service_ruler = nlp.add_pipe("entity_ruler", name="service_ruler", before="ner")
service_ruler.add_patterns(service_patterns)

skill_patterns = load_skill_patterns(jsonl_path)
skill_ruler = nlp.add_pipe("entity_ruler", name="skill_ruler", before="ner")
skill_ruler.add_patterns(skill_patterns)

def test_skill_extraction(sentence):
    doc = nlp(sentence)

    skills = []
    services = []

    for ent in doc.ents:
        if ent.label_ == "SKILLS":
            skills.append(ent.text)
        elif ent.label_ == "SERVICES":
            services.append(ent.text)
    return skills, services

test_sentences = [
    "Experience plumber with expertise in toilet installation, drain unclogging, and pipe repair.",
    "Skilled in electrical wiring, conduit installation, and LED light installation.",
    "Appliance maintenance specialist for washing machine, blenders, and rice cookers.",
    "Furniture repair expert with wood polishing and cushion replacement skills.",
    "I have a degree in electrical enginering, I spent 3 years of work in Company X working as an electrian mainly doing electrical maintenance. I am pretty knowledgable in installing conduits.",
    "Experience plumber with expertise installed a toilet, drain unclogging, and pipe repair."
]

print("Testing Skill Extraction:")
print("=" * 30)

for i, sentence in enumerate(test_sentences, 1):
    extracted_skills, extracted_services = test_skill_extraction(sentence)
    print(f"\nTest {i}:")
    print(f"Sentence: {sentence}")
    print(f"Extracted Services: {extracted_services}")
    print(f"Extracted Skills: {extracted_skills}")
    print("-" * 30)
```
